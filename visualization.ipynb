```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# df = pd.read_csv("datasets/nasa/RW_24.txt", header=None)
df = pd.read_csv("datasets/oxford/q_curve_28_419_cell_1.txt", header=None)
print(df.iloc[0])
```

```python
length = 0
max_value = 0
for i, row in df.iterrows():
    plt.plot(row)
    max_value = max(np.max(row), max_value)
    length = len(row)

plt.plot(np.arange(0, length, 1), np.arange(0, max_value, max_value / length))
# plt.plot(np.arange(0, length, 1), np.arange(0, max_value, 1))

plt.show()
```
